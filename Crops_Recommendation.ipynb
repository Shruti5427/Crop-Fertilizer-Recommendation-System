```python
# Function to extract feature columns
def get_features(df, target_column='label'):
    return [col for col in df.columns if col != target_column]
features = get_features(crop)
print("\nFeature Columns:", features)
```

```python
# Visualizing Histograms for each feature
fig, ax = plt.subplots(3, 3, figsize=(15, 15))
for i, subplot in zip(features, ax.flatten()):
    sns.histplot(crop[i], ax=subplot, kde=True, bins=30)
    subplot.set_title(f"Distribution of {i}")
plt.tight_layout()
plt.show()
```

```python
# Convert categorical label into numeric values for visualization

crop_dict = {label: idx for idx, label in enumerate(crop['label'].unique())}  # Create mapping dictionary
crop['crop_no'] = crop['label'].map(crop_dict)  # Apply mapping
```

```python
# Scatter Plots: Feature vs Crop Type
fig, ax = plt.subplots(3, 3, figsize=(15, 15))
for i, subplot in zip(features, ax.flatten()):
    sns.scatterplot(x=crop[i], y=crop['crop_no'], ax=subplot)
    subplot.set_title(f"Scatter Plot: {i} vs Crop Type")
plt.tight_layout()
plt.show()
```

```python
# Boxplot Visualization for Outlier Detection
fig, ax = plt.subplots(3, 3, figsize=(15, 15))
for i, subplot in zip(features, ax.flatten()):
    sns.boxplot(x=crop[i], ax=subplot)  # Fixed syntax issue
    subplot.set_title(f"Box Plot of {i}")
plt.tight_layout()
plt.show()
```

```python
# Compute correlation matrix for numerical columns
num_cols = crop.select_dtypes(include=[np.number])
corr = num_cols.corr()
```

```python
# Display correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```
